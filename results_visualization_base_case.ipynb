```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('simulation_1.csv', index_col=0)
df.columns
```

```python
plt.figure(figsize=(20, 5))
plt.plot(df['Adults'], label='Adults')
plt.plot(df['Children'], label='Children')
plt.plot(df['Population'], label='Population')
plt.legend()
plt.xlim(0, 500)
```

```python
plt.figure(figsize=(20, 5))
plt.plot(df['Adults']/df['Children'], label='Adults')
plt.legend()
plt.xlim(0, 500)
```

```python
start = 0
end = 500
plt.figure(figsize=(20, 5))
plt.plot(df['Hunger Deaths'].diff().rolling(10).sum()[start:end], label='Hunger Deaths')
plt.plot(df['Old Age Deaths'].diff().rolling(10).sum()[start:end], label='Old Age Deaths')

plt.legend()
plt.xlim(start, end)
```
